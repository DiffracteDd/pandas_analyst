```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
```

```python
y = np.sin(x)
z=np.cos(x)
```

```python
# plotting the sine wave
plt.plot(x,y)
```

```python
# plotting the cos wave
plt.plot(x,z)
```

```python
# adding labels to x axis and y axis
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('Sine wave')
plt.show()
```

```python
# parabola 
x = np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
```

```python
plt.plot(x,y,'r+')
```

```python
plt.plot(x,y,'g-')
```

```python
plt.plot(x,y,'rx')
```

```python
x = np.linspace(-5,5,50)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r--')
```

```python
# Bar plot
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people=[100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('Number Of People')
plt.show()
```

```python
# Pie Chart
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English','French','Spanish','Latin','German']
people = [100,50,150,40,70]
ax.pie(people,labels = languages,autopct='%1.1f%%')
plt.show()
```

```python
#Scatter Plot
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='g')
ax.scatter(x,z,color='r')
plt.show()
```

```python
#3d scatter plot
```

```python
fig3 = plt.figure()
ax = plt.axes(projection='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,z,c=z,cmap='Blues')
plt.show()
```
